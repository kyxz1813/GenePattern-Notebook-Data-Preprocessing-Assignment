```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Question 1.

```python
def gene_file(file, summary = False):
    df = pd.read_csv(file, sep='\t', skiprows=2, index_col = ['Name', 'Description'])
    if summary:
        print('Rows:', df.shape[0])
        print('Columns:', df.shape[1])
    return df
```

### Question 2a.

```python
df_2a = gene_file('https://datasets.genepattern.org/data/test_data/BRCA_minimal_60x19.gct', True)
df_2a
```

<br>

### Question 2b.

```python
df_2b = gene_file('https://datasets.genepattern.org/data/test_data/BRCA_large_20783x40.gct')
df_2b 
```

<br>

### Question 3.

```python
columns = ['A7-A0DB-normal', 'A7-A13E-normal', 'BH-A0B3-primary', 'BH-A0B5-primary']
for col in columns:
    df_3 = df_2b.loc[df_2b[col] < 50000, col]
    plt.figure(figsize=(8, 6))
    plt.hist(df_3, bins=20, edgecolor='black')  
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

<br>

### Question 4.

```python
df_4_index = df_2b.index.get_level_values(0)
df_4_mean = list(df_2b.mean(axis = 1))
df_4_median = list(df_2b.median(axis = 1))
df_4_std = list(df_2b.std(axis = 1))
df_4 = pd.DataFrame(data = {'Sample name': df_4_index, 'Mean': df_4_mean, 'Median': df_4_median, 'Standard deviation': df_4_std})
df_4
```

<br>

### Question 5a.

```python
df_5a = df_2b.loc[df_2b.max(axis = 1) >= 1000]
df_5a
```

<br>

### Question 5b.

```python
columns = ['A7-A0DB-normal', 'A7-A13E-normal', 'BH-A0B3-primary', 'BH-A0B5-primary']
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes = axes.flatten()

for ax, col in zip(axes, columns):
    graph = df_5a.loc[df_5a[col] < 50000, col]
    ax.hist(graph, bins=20, edgecolor='black')
    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
